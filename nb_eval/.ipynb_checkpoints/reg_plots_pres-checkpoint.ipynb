```python
mympf.setMatplotlibParam_KdePlot()
```

```python
axs = mympf.set_figure_axs(
    1,
    1,
    wratio=0.35,
    hratio=0.75,
    pad_w_ext_left=0.1,
    pad_w_ext_right=0.1,
    pad_w_int=0.001,
    pad_h_ext=0.001,
    pad_h_int=0.25,
)
ax = axs[0]
sns.kdeplot(
    data=df_mape,
    x=first_col,
    common_norm=True,
    hue=second_col,
    fill=True,
    alpha=0.2,
    ax=ax,
)
ax.set_xticks([0, 25, 50, 75, 100])
ax.set_xlim(0, 100)
ax.set_yticks([])
ax.set_yticklabels([])
```
