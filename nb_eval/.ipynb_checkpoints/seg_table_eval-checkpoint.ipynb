```python
def eval_ds_with_model(data, model):
    x_valid = tf.convert_to_tensor(data.x.valid[0], np.float32)
    y_valid = tf.convert_to_tensor(data.y.valid, np.float32)
    loss_valid = model.evaluate(x=x_valid, y=y_valid, batch_size=64)
    return loss_valid
```
