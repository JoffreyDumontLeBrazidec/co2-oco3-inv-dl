## Histogram

```python
df_mae_1 = pd.DataFrame({"loss": metrics["mae"], "method": "CNN"})
df_mae_2 = pd.DataFrame({"loss": mean_metrics["mae"], "method": "mean"})
df_mae = pd.concat([df_mae_1, df_mae_2])
```

```python
df_mape_1 = pd.DataFrame({"loss": metrics["mape"], "method": "CNN"})
df_mape_2 = pd.DataFrame({"loss": mean_metrics["mape"], "method": "mean"})
df_mape = pd.concat([df_mape_1, df_mape_2])
```

```python
pred = np.squeeze(tf.convert_to_tensor(
    model.predict(tf.convert_to_tensor(data.x.eval, np.float32)), np.float32
).numpy())
y = data.y.eval[:, -1]
df_emiss_1 = pd.DataFrame({"emiss": y, "origin": "truth"})
df_emiss_2 = pd.DataFrame({"emiss": pred, "origin": "prediction"})
df_emiss = pd.concat([df_emiss_1, df_emiss_2])
```

```python
N_rows = 3
N_cols = 2
matplotlib_functions.setMatplotlibParam()
plt.viridis()
axs = matplotlib_functions.set_figure_axs(
    N_rows,
    N_cols,
    wratio=0.35,
    hratio=0.75,
    pad_w_ext_left=0.25,
    pad_w_ext_right=0.25,
    pad_w_int=0.3,
    pad_h_ext=0.2,
    pad_h_int=0.35,
)

sns.kdeplot(
    data=df_mae,
    x="loss",
    common_norm=True,
    hue="method",
    color="firebrick",
    fill=True,
    alpha=0.2,
    ax=axs[0],
)
sns.kdeplot(
    data=df_mape,
    x="loss",
    common_norm=True,
    hue="method",
    color="firebrick",
    fill=True,
    alpha=0.2,
    ax=axs[1],
)
sns.kdeplot(
    data=df_emiss,
    x="emiss",
    common_norm=True,
    hue="origin",
    color="firebrick",
    fill=True,
    alpha=0.2,
    ax=axs[2],
)
sns.kdeplot(pred/y, color="firebrick", fill=True, alpha=0.2, ax=axs[3])
sns.kdeplot(pred-y, color="firebrick", fill=True, alpha=0.2, ax=axs[4])
sns.kdeplot(np.abs(pred)-np.abs(y), color="firebrick", fill=True, alpha=0.2, ax=axs[5])

titles = [
    "Mean absolute error",
    "Mean absolute percentage error",
    "emiss",
    "pred/truth",
    "pred-truth",
    "abs(pred-truth)",
]

for i_ax, ax in enumerate(axs):
    ax.set_yticklabels([])
    ax.set_xlabel("")
    ax.set_xlabel(titles[i_ax])
    #ax.set_title(titles[i_ax])
    #ax.text(.01, .99, titles[i_ax], ha='left', va='top', transform=ax.transAxes, fontsize=8)
```
