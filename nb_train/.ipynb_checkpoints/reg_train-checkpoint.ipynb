### Import lib

```python
import os
os.chdir("/cerea_raid/users/dumontj/dev/coco2/dl")
import numpy as np
import sys
import pandas as pd
import xarray as xr
import tensorflow as tf
from tensorflow import keras
import keras.backend as K
import matplotlib.pyplot as plt
import seaborn as sns
from treeconfigparser import TreeConfigParser

from include.loss import calculate_weighted_plume

import matplotlib_functions
matplotlib_functions.setMatplotlibParam()
plt.viridis()
import joblib
import pickle

from sklearn import preprocessing
from sklearn.linear_model import LinearRegression
import sklearn.linear_model as linear_model
import sklearn.ensemble as ensemble

from data.Data import Data, Eval_Shuffler
from include.loss import pixel_weighted_cross_entropy
import wandb
from wandb.keras import WandbCallback
import models.reg as rg 
from include.loss import rmae

dir_res = "/cerea_raid/users/dumontj/dev/coco2/dl/res"
dir_plots = "/cerea_raid/users/dumontj/dev/coco2/dl/eval/reg"
dir_eval_ddeq = "/libre/dumontj/coco2/ddeq/eval"
```

```python
def get_data(name_ds: str, scaler, mode: str):
    config = TreeConfigParser()
    config.readfiles(os.path.join(dir_res, name_ds, "config.cfg"))
    with open(os.path.join(dir_res, name_ds, "tv_inds.pkl"), "rb") as handle:
        tv_inds = pickle.load(handle)
    data = Data(config, tv_inds, mode)
    data.prepare_input(scaler)
    data.prepare_output()
    return data


def get_data_name(weights: str):
    config_weights = TreeConfigParser()
    config_weights.readfiles(os.path.join(dir_res, weights, "config.cfg"))
    weights_ds = config_weights.get("data.directory.name")
    return weights_ds

def get_model(path_w: str):
    model = keras.models.load_model(os.path.join(path_w, "weights_cp_best.h5"),
                                    custom_objects={"relative_mae": rmae, "rmae": rmae},
                                    compile=True)
    return model

def get_seg_model(path_w: str):
    model = keras.models.load_model(os.path.join(path_w, "weights_cp_best.h5"), compile=False)
    model.compile('adam', loss=pixel_weighted_cross_entropy)    
    return model

def get_scaler(path_w: str):
    scaler = joblib.load(os.path.join(path_w, "scaler.save"))
    return scaler

mode = "train"
```

```python
generator = keras.preprocessing.image.ImageDataGenerator(
            rotation_range=180,
            width_shift_range=0.2,
            height_shift_range=0.2,
            horizontal_flip=True,
            vertical_flip=True,
            shear_range=90,
            zoom_range=0.2,
        )
```

## PLUME

```python
dataset = "2km_S_pBer"
dir_dataset = os.path.join("/libre/dumontj/coco2/dl-input", dataset)
ds = xr.open_dataset(os.path.join(dir_dataset, "2d_train_valid_dataset.nc"))
path_w = os.path.join("nb_train/sim/weights", dataset)
```

```python
x = np.expand_dims(ds.plume.values, -1)
y = ds.emiss.values[:, 1]
y_full = ds.emiss.values
N_data = ds.attrs["N_img"]

with open(os.path.join(dir_dataset, "tv_inds.pkl"), "rb") as handle:
    tv_inds = pickle.load(handle)
eval_shuffler = Eval_Shuffler(
    train_ratio=0.8,
    N_data=N_data,
    ds_inds=tv_inds,
)

[x_train, x_valid] = eval_shuffler.train_valid_split(x)
[y_train, y_valid] = eval_shuffler.train_valid_split(y)
[y_full_train, y_full_valid] = eval_shuffler.train_valid_split(y_full)

#x_scaler = preprocessing.StandardScaler()
#x_train = x_scaler.fit_transform(x_train.reshape(-1, x_train.shape[-1]))
#x_valid = x_scaler.transform(x_valid.reshape(-1, x_valid.shape[-1])).reshape(x_valid.shape)
```

### Model def

```python
model_builder = rg.Keras_reg_model_builder("EfficientNetB0", x.shape[1:], 1)
model = model_builder.get_model()
model.compile("adam", loss=tf.keras.losses.MeanAbsoluteError())
```

### Training

```python
modelcheckpoint_cb = keras.callbacks.ModelCheckpoint(
    filepath=os.path.join(path_w, "weights_plume_to_emiss.h5"),
    save_weights_only=False,
    monitor="val_loss",
    save_best_only=True,
    mode="auto",
    verbose=1,
)
```

```python
model.fit(generator.flow(x_train, y_train), epochs=1000, callbacks = [modelcheckpoint_cb], 
          validation_data=(x_valid, y_valid))
```

### Evaluation

```python
model = keras.models.load_model(os.path.join(path_w, "weights_plume_to_emiss.h5"), compile=True)
```

```python
x_valid = tf.convert_to_tensor(x_valid, np.float32)
y_valid = tf.convert_to_tensor(np.expand_dims(y_valid, -1), np.float32)

#ds_valid = ds.sel(idx_img=eval_shuffler.ds_inds["valid"])
pred_valid = tf.convert_to_tensor(np.expand_dims(model.predict(x_valid), -1), np.float32)

mape = tf.keras.losses.MeanAbsolutePercentageError(reduction=tf.keras.losses.Reduction.NONE,)
mae = tf.keras.losses.MeanAbsoluteError(reduction=tf.keras.losses.Reduction.NONE,)

losses_mape = np.squeeze(mape(y_valid, pred_valid))
losses_mae = np.squeeze(mae(y_valid, pred_valid))

print("CNN, mape:", np.mean(losses_mape), np.median(losses_mape))
print("CNN, mae:", np.mean(losses_mae), np.median(losses_mae))
```

#### Baseline mean

```python
mean_valid = tf.math.reduce_mean(y_valid) * tf.ones_like(y_valid, np.float32)

mean_baseline_mape = np.squeeze(mape(y_valid, mean_valid))
mean_baseline_mae = np.squeeze(mae(y_valid, mean_valid))

print("MEAN, mape:", np.mean(mean_baseline_mape), np.median(mean_baseline_mape))
print("MEAN, mae:", np.mean(mean_baseline_mae), np.median(mean_baseline_mae))
```

#### Baseline linear regression (plume mass -> emissions)

```python
linear_reg = LinearRegression().fit(np.sum(x_train, axis=(1,2)), y_train)
linear_valid = linear_reg.predict(np.sum(x_valid, axis=(1,2)))

linear_baseline_mape = np.squeeze(mape(y_valid, linear_valid))
linear_baseline_mae = np.squeeze(mae(y_valid, linear_valid))

print("LINEAR, mape:", np.mean(linear_baseline_mape), np.median(linear_baseline_mape))
print("LINEAR, mae:", np.mean(linear_baseline_mae), np.median(linear_baseline_mae))
```

#### Histogram perf.

```python
df_mape = pd.DataFrame({'loss': losses_mape,
                      'method': ["CNN"]*len(losses_mape) })
df_mape = pd.concat([df_mape, pd.DataFrame({'loss': mean_baseline_mape,
                       'method': ["mean"]*len(mean_baseline_mape) })], ignore_index=True)
df_mape = pd.concat([df_mape, pd.DataFrame({'loss': np.squeeze(linear_baseline_mape),
                       'method': ["linear"]*len(linear_baseline_mape) })], ignore_index=True)

df_mae = pd.DataFrame({'loss': losses_mae,
                      'method': ["CNN"]*len(losses_mae) })
df_mae = pd.concat([df_mae, pd.DataFrame({'loss': mean_baseline_mae,
                       'method': ["mean"]*len(mean_baseline_mae) })], ignore_index=True)
df_mae = pd.concat([df_mae, pd.DataFrame({'loss': np.squeeze(linear_baseline_mae),
                       'method': ["linear"]*len(linear_baseline_mae) })], ignore_index=True)

df_emiss = pd.DataFrame({'val': np.squeeze(y_valid.numpy()),
                      'origin': ["truth"]*len(np.squeeze(y_valid.numpy())) })
df_emiss = pd.concat([df_emiss, pd.DataFrame({'val': np.squeeze(pred_valid.numpy()),
                       'origin': ["pred"]*len(np.squeeze(pred_valid.numpy())) })], ignore_index=True)
```

```python
N_rows=3
N_cols=2
matplotlib_functions.setMatplotlibParam()
plt.viridis()
axs = matplotlib_functions.set_figure_axs(N_rows, N_cols, wratio=0.35, hratio=0.75, pad_w_ext_left=0.25, 
                                          pad_w_ext_right=0.25, pad_w_int=0.3, pad_h_ext=0.2, pad_h_int=0.35)

sns.kdeplot(data=df_mae, x="loss", common_norm=True, hue="method", color="firebrick", fill=True, alpha=0.2, ax=axs[0])
sns.kdeplot(data=df_mape, x="loss", common_norm=True, hue="method", color="firebrick", fill=True, alpha=0.2, ax=axs[1])
sns.kdeplot(data=df_emiss, x="val", common_norm=True, hue="origin", color="firebrick", fill=True, alpha=0.2, ax=axs[2])
sns.kdeplot(np.squeeze(pred_valid.numpy())/np.squeeze(y_valid.numpy()), color="firebrick", fill=True, alpha=0.2, ax=axs[3])
sns.kdeplot(np.squeeze(pred_valid.numpy())-np.squeeze(y_valid.numpy()), color="firebrick", fill=True, alpha=0.2, ax=axs[4])
sns.kdeplot(np.abs(np.squeeze(pred_valid.numpy())-np.squeeze(y_valid.numpy())), color="firebrick", fill=True, alpha=0.2, ax=axs[5])

titles = ["Mean absolute error", "Mean absolute percentage error", "emiss", "pred/truth", "pred-truth", "abs(pred-truth)"]

for i_ax, ax in enumerate(axs):
    ax.set_yticklabels([])
    ax.set_xlabel("")
    ax.set_title(titles[i_ax])
```
