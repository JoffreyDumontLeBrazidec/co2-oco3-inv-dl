### Data download

```python
path_res = "/cerea_raid/users/dumontj/dev/coco2/dl/res"
dataset = "seg_PGIPW"
config = TreeConfigParser()
config.readfiles(os.path.join(path_res, dataset, "config.cfg"))
shuffle_indices = np.fromfile(os.path.join(path_res, dataset, "shuffle_indices.bin")).astype(np.int64)
data = Data(config, shuffle_indices)
data.prepare_input()
data.prepare_output()
```

### Generator

```python
generator = keras.preprocessing.image.ImageDataGenerator(
#            rotation_range=180,
            width_shift_range=0,
            height_shift_range=0,
#            horizontal_flip=True,
#            vertical_flip=True,
            shear_range=0,
            zoom_range=0,
        )
```

### Model def

```python
model = sm.Unet_efficient(data.x.fields_input_shape, data.y.classes)
model.compile('adam', loss=pixel_weighted_cross_entropy)
```

### Training

```python
modelcheckpoint_cb = keras.callbacks.ModelCheckpoint(
    filepath=os.path.join("nb/sim/weights", dataset + ".h5"),
    save_weights_only=False,
    monitor="val_loss",
    mode="max",
    save_best_only=True
)
callbacks=[modelcheckpoint_cb]
```

```python
wandb.init()
callbacks=[modelcheckpoint_cb, WandbCallback()]
```

```python
model.fit(generator.flow(data.x.train[0], data.y.train), 
          epochs=50, validation_data=(data.x.valid, data.y.valid), 
          callbacks=callbacks)
```

```python
model.fit(x=data.x_train, y=data.y_train, epochs=10, validation_data=(data.x_valid, data.y_valid), 
          callbacks=[modelcheckpoint_cb])
```

### Save and load model

```python
model.save(os.path.join("nb/sim/weights", dataset + ".h5"))
```

```python
model = keras.models.load_model(os.path.join("nb/sim/weights", dataset + ".h5"), compile=False)
model.compile('adam', loss=pixel_weighted_cross_entropy)
```

## Evaluation

```python
x_valid = tf.convert_to_tensor(data.x_valid[0], np.float32)
pred_valid = tf.convert_to_tensor(model.predict(x_valid), np.float32)
y_valid = tf.convert_to_tensor(data.y_valid, np.float32)
```

```python
loss_valid = model.evaluate(x=x_valid, y=y_valid, batch_size=64)
```

### Random output

```python
idx = int(np.random.uniform(0,data.y_valid.shape[0]*0.2))
print(model.compute_loss(y=y_valid[idx:idx+1], y_pred=pred_valid[idx:idx+1]).numpy())

plt.figure(figsize=(12,10))

plt.subplot(2, 2, 1)
plt.imshow(np.squeeze(data.trace_valid[idx]))
plt.colorbar()
plt.title("trace")
#plt.savefig("nb/sim/fig/seg_pw-trace.png", transparent=True)

plt.subplot(2, 2, 2)
plt.imshow(np.squeeze(data.y_valid[idx]))
plt.colorbar()
plt.title("truth")
#plt.savefig("nb/sim/fig/seg_pw-truth.png")

plt.subplot(2, 2, 3)
plt.imshow(np.squeeze(data.x_valid[0][idx]))
plt.colorbar()
plt.title("image")
#plt.savefig("nb/sim/fig/seg_pw-input.png")

plt.subplot(2, 2, 4)
plt.imshow(np.squeeze(model(tf.expand_dims(data.x_valid[0][idx], 0))[0]))
plt.colorbar()
plt.title("pred")
#plt.savefig("nb/sim/fig/seg_pw-prediction.png")
```

```python
print(pixel_weighted_cross_entropy(y_valid,pred_valid,reduction=True))
print(model.compute_loss(y=y_valid, y_pred=pred_valid))
```

```python
loss_valid = pixel_weighted_cross_entropy(y_valid,pred_valid,reduction=False)
loss_valid = np.mean(loss_valid,axis=(1,2))
```

### Worst evaluation

```python
idx_worst = np.argmax(loss_valid)
print(model.compute_loss(y=y_valid[idx_worst:idx_worst+1], y_pred=pred_valid[idx_worst:idx_worst+1]))

plt.figure(figsize=(12,10))

plt.subplot(2, 2, 1)
plt.imshow(np.squeeze(data.trace_valid[idx_worst]))
plt.colorbar()
plt.title("trace")
#plt.savefig("nb/sim/fig/seg_pw-trace.png", transparent=True)

plt.subplot(2, 2, 2)
plt.imshow(np.squeeze(data.y_valid[idx_worst]))
plt.colorbar()
plt.title("truth")
#plt.savefig("nb/sim/fig/seg_pw-truth.png")

plt.subplot(2, 2, 3)
plt.imshow(np.squeeze(data.x_valid[0][idx_worst]))
plt.colorbar()
plt.title("image")
#plt.savefig("nb/sim/fig/seg_pw-input.png")

plt.subplot(2, 2, 4)
plt.imshow(np.squeeze(model(tf.expand_dims(data.x_valid[0][idx_worst], 0))[0]))
plt.colorbar()
plt.title("pred")
#plt.savefig("nb/sim/fig/seg_pw-prediction.png")

```

### Best evaluation

```python
idx_best = np.argmin(loss_valid)

plt.figure()
plt.imshow(np.squeeze(data.trace_valid[idx_best]))
plt.colorbar()
plt.title("trace")
plt.savefig("nb/sim/fig/seg_pw-trace.png", transparent=True)

plt.figure()
plt.imshow(np.squeeze(data.y_valid[idx_best]))
plt.colorbar()
plt.title("truth")
plt.savefig("nb/sim/fig/seg_pw-truth.png")

plt.figure()
plt.imshow(np.squeeze(model(tf.expand_dims(data.x_valid[0][idx_best], 0))[0]))
plt.colorbar()
plt.title("pred")
plt.savefig("nb/sim/fig/seg_pw-prediction.png")

plt.figure()
plt.imshow(np.squeeze(data.x_valid[0][idx_best]))
plt.colorbar()
plt.title("image")
plt.savefig("nb/sim/fig/seg_pw-input.png")
```
