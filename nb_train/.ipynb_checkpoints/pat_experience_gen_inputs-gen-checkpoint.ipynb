### Import lib

```python
import os
os.chdir("/cerea_raid/users/dumontj/dev/coco2/dl")
import numpy as np
import sys
import pandas as pd
import xarray as xr
import tensorflow as tf
from tensorflow import keras
import keras.backend as K
import matplotlib.pyplot as plt
import seaborn as sns
from treeconfigparser import TreeConfigParser

import models.reg as rg 
from data.Data import Data
```

### Data generation

```python
#"reg_pw", "reg_par", "reg_ber", 
#"reg_PGIPW", "reg_PGIBJB", "reg_PGIBLB", "reg_BPW"
#"reg_PGIBPW"
dataset = "reg_par"
config_file = os.path.join("nb/sim/cfg", dataset + ".cfg")
config = TreeConfigParser()
config.readfiles(config_file)

data = Data(config)
data.prepare_input()
data.prepare_output()
```

### Model def

```python
model_builder = rg.Keras_reg_model_builder("EfficientNetB0", data.x.fields_input_shape, data.y.classes)
model = model_builder.get_model()
model.compile("adam", loss=tf.keras.losses.MeanSquaredLogarithmicError())
```

```python
model = rg.classic_CNN("EfficientNetB0", data.x.fields_input_shape, data.y.classes, None)
model.compile("adam", loss=tf.keras.losses.MeanSquaredLogarithmicError())
```

### Training

```python
modelcheckpoint_cb = keras.callbacks.ModelCheckpoint(
    filepath=os.path.join("nb/sim/weights", dataset + ".h5"),
    save_weights_only=False,
    monitor="val_loss",
    mode="max",
    save_best_only=True
)
```

```python
model.fit(x=data.x.train, y=data.y.train, epochs=30, callbacks = [modelcheckpoint_cb], 
          validation_data=(data.x.valid, data.y.valid))
```

### Save and load model

```python
model.save(os.path.join("nb/sim/weights", dataset + ".h5"))
```

```python
model = keras.models.load_model(os.path.join("nb/sim/weights", dataset + ".h5"), compile=True)
```

## Evaluation

```python
x_valid = tf.convert_to_tensor(data.x.valid[0], np.float32)
pred_valid = tf.convert_to_tensor(model.predict(x_valid), np.float32)
y_valid = tf.convert_to_tensor(data.y.valid, np.float32)
```

```python
idx = int(np.random.uniform(0,data.y.valid.shape[0]*0.2))
print(model.compute_loss(y=y_valid[idx:idx+1], y_pred=pred_valid[idx:idx+1]).numpy())

print("truth:", np.squeeze(data.y.valid[idx]))

print("pred:", np.squeeze(model(tf.expand_dims(data.x.valid[0][idx], 0))[0]))

plt.figure()
plt.imshow(np.squeeze(data.x.valid[0][idx,:,:,0]))
plt.colorbar()
plt.title("image")
plt.savefig("nb/sim/fig/reg-truth.png")
```

### Histogram perf.

```python
df = pd.DataFrame({'y_true': np.squeeze(data.y.valid), 'y_pred': np.squeeze(pred_valid.numpy()), 
                   'pred/truth': np.squeeze(pred_valid.numpy())/np.squeeze(data.y.valid)})
df["y_true_bin"] = pd.qcut(df["y_true"], q=4, precision=2)
plt.figure(figsize=(15,8))
sns.histplot(
    data=df,
    x="pred/truth",
    hue="y_true_bin",
    log_scale=False,
    element="step",
    #cumulative=True,
    stat="density",
    common_norm=False,
    bins=200,
)
```
